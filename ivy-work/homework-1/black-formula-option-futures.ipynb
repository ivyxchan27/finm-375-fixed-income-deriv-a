# **1. Black’s Formula for Bond Options**
### 1.1.
Consider a bond with:
* T=3
* face value of N=100
* coupons at annual frequency
* annualized coupon rate of cpn=6%.

Use the bond-pricing formula along with the discount rates in the data file to price this bond.

```python
time_to_maturity = 3
face_value = 100.0
cpn = 0.06 #convert percent to decimal

cpn_payment = face_value * cpn  #annual coupon = 6

Z1 = data.loc[data["ttm"] == 1.0, "discount"].item()
Z2 = data.loc[data["ttm"] == 2.0, "discount"].item()
Z3 = data.loc[data["ttm"] == 3.0, "discount"].item()

bond_price = cpn_payment * Z1 + cpn_payment * Z2 + (cpn_payment + face_value) * Z3
bond_price
```

### 1.3.
What is the price of the callable bond?
